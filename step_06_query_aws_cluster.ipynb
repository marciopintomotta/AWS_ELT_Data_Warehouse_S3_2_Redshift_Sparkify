### With the environment configured, we can now query the data.

##### 1. Queries the number of records of the time dimension.

```python
%%sql
SELECT DISTINCT COUNT(*)
FROM dim_time
```

##### 2. Queries the number of artist is null in the stage events table

```python
%%sql
SELECT count(*)  
FROM staging_events     
WHERE artist IS NULL
```

##### 3. Queries the "valid" artist in the stage events table

```python
%%sql
SELECT artist  
FROM staging_events     
WHERE artist IS not NULL
group by artist
order by artist
```

##### 3. Queries the artists that exist in the stage table but do not exist in the dimensional table of the articles.

```python
%%sql
select
 art.artist
from (
    select artist  
    from staging_events     
    where artist IS not NULL
    group by artist
) art
LEFT JOIN dim_artists a ON (art.artist = a.name) 
where
a.name is null
order by art.artist
```

##### 4. Queries the artists in the dimensional table of the articles.

```python
%%sql
select * from dim_artists
order by name
```
